```python
import pandas as pd

# Load the sample CSV
df = pd.read_csv('sample_ai_healthcare_tweets.csv')

# Display the first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Ensure created_at is datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Group by date and sentiment
sentiment_trend = df.groupby([df['created_at'].dt.date, 'sentiment']).size().unstack(fill_value=0)

# Plot
sentiment_trend.plot(kind='line', figsize=(12, 6), marker='o')
plt.title("Sentiment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def top_keywords(df, sentiment, n=10):
    texts = df[df['sentiment'] == sentiment]['text']
    cv = CountVectorizer(stop_words='english', max_features=1000)
    word_counts = cv.fit_transform(texts)
    sum_words = word_counts.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
    sorted_words = sorted(words_freq, key=lambda x: x[1], reverse=True)[:n]

    words, freqs = zip(*sorted_words)
    plt.figure(figsize=(8, 5))
    plt.barh(words[::-1], freqs[::-1], color='skyblue')
    plt.title(f"Top {n} Keywords in {sentiment.capitalize()} Tweets")
    plt.xlabel("Frequency")
    plt.show()

top_keywords(df, 'positive')
top_keywords(df, 'negative')
top_keywords(df, 'neutral')
```

```python
# Load the CSV file
df = pd.read_csv("tweets.csv")

# Show the first few rows
df.head()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Drop rows with invalid/missing dates (optional but safe)
df = df.dropna(subset=['created_at'])
```

```python
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = analyzer.polarity_scores(str(text))['compound']
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['text'].apply(get_sentiment)

df[['text', 'sentiment']].head()
```

```python
sns.countplot(data=df, x='sentiment', palette='pastel')
plt.title("Overall Sentiment Distribution")
plt.grid(True)
plt.show()
```

```python
# Combine text by sentiment
for sentiment in ['positive', 'neutral', 'negative']:
    text = " ".join(df[df['sentiment'] == sentiment]['text'].dropna().astype(str))
    wc = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 4))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"WordCloud for {sentiment} tweets")
    plt.show()
```

```python
from collections import Counter
import re

def clean_text(text):
    return re.sub(r'[^\w\s]', '', str(text).lower())

for sentiment in ['positive', 'neutral', 'negative']:
    words = " ".join(df[df['sentiment'] == sentiment]['text'].dropna()).split()
    words = [clean_text(word) for word in words if len(word) > 3]
    top = Counter(words).most_common(10)

    print(f"\nTop words in {sentiment} tweets:")
    for word, count in top:
        print(f"{word}: {count}")
```

```python
# Grouping only if data exists
if not df.empty and 'created_at' in df.columns:
    # Ensure no missing sentiment or created_at
    df_clean = df.dropna(subset=['sentiment', 'created_at'])

    # Resample and get mode sentiment per day
    df_clean.set_index('created_at', inplace=True)

    daily_sentiment = df_clean.resample('D')['sentiment'].apply(
        lambda x: x.value_counts().idxmax() if not x.empty else None
    ).dropna()

    # Plotting
    plt.figure(figsize=(12, 4))
    daily_sentiment.value_counts().plot(kind='bar', color='lightgreen')
    plt.title("Most Common Sentiment Per Day")
    plt.ylabel("Days Count")
    plt.grid(True)
    plt.show()
else:
    print("No valid data to process.")
```
